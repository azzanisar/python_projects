```python
from newspaper import Article
from textblob import TextBlob

def analyze_sentiment(text):
    """
    Makes sentiment analysis for the given text
    """
    # Makes sentiment analysis on the text.
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity

    # Print the analysis results.
    if polarity > 0 and subjectivity < 0.5:
        result = "Positive and objective"
    elif polarity > 0 and subjectivity > 0.5:
        result = "Positive and subjective"
    elif polarity > 0 and subjectivity == 0.5:
        result = "Positive and neutral content"
    elif polarity < 0 and subjectivity < 0.5:
        result = "Negative and objective"
    elif polarity < 0 and subjectivity > 0.5:
        result = "Negative and subjective"
    elif polarity < 0 and subjectivity == 0.5:
        result = "Negative and neutral content"
    elif polarity == 0 and subjectivity < 0.5:
        result = "Neutral and objective"
    elif polarity == 0 and subjectivity > 0.5:
        result = "Neutral and subjective"
    elif polarity == 0 and subjectivity == 0.5:
        result = "Neutral"    
    else:
        result = "Not able to calculate sentiment of this content! Try another input"

    print("Polarity: ", polarity)
    print("Subjectivity: ", subjectivity)
    print("News status: ", result)
    
    return polarity, subjectivity


def get_article_sentiment():
    """
    Make sentiment analysis for the given news link
    """
    # Get a news link input from the user
    url = input("Please enter a news link: ")

    # Extract the news content using newspaper3k
    article = Article(url)
    article.download()
    article.parse()
    text = article.text
    
    polarity, subjectivity = analyze_sentiment(text)

    # Check the news status
    if polarity > 0 and subjectivity < 0.5:
        result = "Positive and objective"
    elif polarity > 0 and subjectivity > 0.5:
        result = "Positive and subjective"
    elif polarity > 0 and subjectivity == 0.5:
        result = "Positive and neutral content"
    elif polarity < 0 and subjectivity < 0.5:
        result = "Negative and objective"
    elif polarity < 0 and subjectivity > 0.5:
        result = "Negative and subjective"
    elif polarity < 0 and subjectivity == 0.5:
        result = "Negative and neutral content"
    elif polarity == 0 and subjectivity < 0.5:
        result = "Neutral and objective"
    elif polarity == 0 and subjectivity > 0.5:
        result = "Neutral and subjective"
    elif polarity == 0 and subjectivity == 0.5:
        result = "Neutral"    
    else:
        result = "Not able to calculate sentiment of this content! Try another input"

    # Print the analysis results.
    print("News content:")
    print(text)
    print("Polarity: ", polarity)
    print("Subjectivity: ", subjectivity)
    print("News status: ", result)
    
    return polarity, subjectivity


def get_text_sentiment():
    """
    Sentiment analysis for the given text.
    """
    text = input("Please enter a news content: ")
    return analyze_sentiment(text)


def main():
    polarities = []
    subjectivities = []
    while True:
        choice = input("1. Enter a text\n2. Enter a link\n3. Exit\nChoice: ")
        if choice == "1":
            polarity, subjectivity = get_text_sentiment()
            polarities.append(polarity)
            subjectivities.append(subjectivity)
            print("Sentiment: polarity = {}, subjectivity = {}".format(polarity, subjectivity))
        elif choice == "2":
            polarity, subjectivity = get_article_sentiment()
            polarities.append(polarity)
            subjectivities.append(subjectivity)
            print("Sentiment: polarity = {}, subjectivity = {}".format(polarity, subjectivity))
        elif choice == "3":
            avg_polarity = sum(polarities) / len(polarities) if len(polarities) > 0 else 0
            avg_subjectivity = sum(subjectivities) / len(subjectivities) if len(subjectivities) > 0 else 0
            print("Average polarity: {}".format(avg_polarity))
            print("Average subjectivity: {}".format(avg_subjectivity))
            break
        else:
            print("Invalid choice.")


if __name__ == '__main__':
    main()
```
